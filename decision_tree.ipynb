```python
# your task for tomorrow is to create a gradio application on top of any of these.
# create an end to end application using tips data set and gradio
# Import the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
numerical_columns=["total_bill","size"]
categorical_columns=["sex","smoker","day","time"]
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),categorical_columns)],remainder="passthrough")
```

```python
# separate independent variable and a dependent variable which is a tip column
X=df.drop(columns=["tip"])
y=df["tip"]
```

```python
# split the dataset into train and test # hold out method
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# create the pipeline for this, which includes steps for preprocessing using column transformer and  randomforest regressor
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
pipe=Pipeline(steps=[("preprocessor",ct),("model",RandomForestRegressor())])
```

```python
# Training
pipe.fit(X_train,y_train)
```

```python
# gradio application

def prediction(total_bill,sex,smoker,day,time,size):
  inputs_df=pd.DataFrame([[total_bill,sex,smoker,day,time,size]],columns=["total_bill","sex","smoker","day","time","size"])
  # print(inputs_df)
  output=pipe.predict(inputs_df)
  # print(output)
  return output[0]
```

```python
%%capture
pip install gradio
```

```python
# create gradio application and add dropdown for categorical variables
import gradio as gr
total_bill_input=gr.Number(label="Total Bill Amount")
gender_menu=gr.Dropdown(label="sex",choices=["Male","Female"])
smoker_menu=gr.Dropdown(label="smoker",choices=["Yes","No"])
day_menu=gr.Dropdown(label="day",choices=["Sat","Sun","Fri","Thur"])
time_menu=gr.Dropdown(label="time",choices=["Lunch","Dinner"])
size_input=gr.Number(label="Size of the family")
```

```python
demo=gr.Interface(fn=prediction,inputs=[total_bill_input,gender_menu,smoker_menu,day_menu,time_menu,size_input],
                  outputs=gr.Num    ber(label="Predicted tip amount"))
```

```python
demo.launch(share=True)
```
